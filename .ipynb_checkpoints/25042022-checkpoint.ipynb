```python
mean = [0, 0] # wartość średnia 
cov = [[1, 1], [1, 2]]
# macierz kowariancji, jest ton uogólnienienie wariancji na przypadek wielowymiarowy
x, y = np.random.multivariate_normal(mean, cov, 10000).T # do ML się przyda duuużo algebry
# .T - transpozycja macierzy, czyli odwrócenie macierzy
# .random.multivariate_normal - losowe próbki z wielowymiarowego rozkładu normalnego
```

```python
plt.figure(figsize=(10, 6), dpi=120)
plt.hist2d(x, y, bins=30, cmap='Blues') #hist2d - dwuwymiarowy (2D) histogram
cb = plt.colorbar() # plt.colorbar dodaje pasek kolorów po prawo
cb.set_label('counts in bin')
```

```python
plt.figure(figsize=(10, 6), dpi=120)
plt.hexbin(x, y, gridsize=50, cmap='Blues')
# sześciokątny wykres dwuwymiarowy punktów x, y
# gridsize określa liczbę sześciokątów w kierunku x
cb = plt.colorbar(label='count in bin')
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births['decade'] = 10 * (births['year'] // 10)
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum') # działa trochę jak groupby. grupujemy w tym wypadku
# po kolumnie decada, zliczamy to co jest w births (tutaj zadeklarowaliśmy że ma zrobić sum, dodatkowo grupujemy po płci, taka transpozycja z kolumny gender)
```

```python
births.head()
```

```python
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year');
```

```python
births.info()
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75]) # oblicza q-ty percentyl z danych wzdłuż podanej osi
# percentyl to kwantyl rzędu k/100, gdzie k = 1, ..., 99
# Percentyl jest wielkością, poniżej której padają wartości zadanego procentu próbek
mu = quartiles[1] # wartość oczekiwana
sig = 0.74 * (quartiles[2] - quartiles[0]) # sigma
```

```python
births = births.dropna()
```

```python
births['day'] = births['day'].astype(int)
# astype tworzy kopie talicy i rzutuje ją na określony typ
```

```python
births['day'] = births['day'].astype(int)
# astype tworzy kopie talicy i rzutuje ją na określony typ
```

```python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
```

```python
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day,
                              format='%Y%m%d')
births['dayofweek'] = births.index.dayofweek
```

```python
births.head()
```

```python
births.pivot_table('births', index='dayofweek',columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
```

```python
births_by_date = births.pivot_table(values='births', index=[births.index.month, births.index.day])
births_by_date.head()
```

```python
from datetime import datetime

births_by_date.index = [datetime(2012, month, day)
                        for (month, day) in births_by_date.index]
births_by_date.head()
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

style = dict(size=10, color='gray')

ax.text(x='2012-1-1', y=3950, s="New Year's Day", **style)
ax.text(x='2012-7-4', y=4250, s="Independence Day", ha='center', **style) # ha - horizontal alignment
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)

# Oznaczenia osi
ax.set(title='USA births by day of year (1969-1988)',
       ylabel='average daily births');
```

```python
from datetime import timedelta
```

```python
def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return timedelta(hours=h, minutes=m, seconds=s)
```

```python
marathon_df = pd.read_csv(marathon_url, converters={'split':convert_time, 'final':convert_time})
```

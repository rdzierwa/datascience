```python
import sqlite3
```

```python
conn = sqlite3.connect('exercise_01')
```

```python
for filename in filenames:
    print(f"Loading {filename}...")
    skiprows = 54 if filename == 'flights' else 0
    df = pd.read_csv(f"Exercise01/{filename}.csv", skiprows=skiprows)
    df.to_sql(filename, con=conn, if_exists='replace')
    dfs[filename] = df
    print('-'*10)
```

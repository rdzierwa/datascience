# Srcaping Media Bias Ratings

```python
r = requests.get('https://www.allsides.com/media-bias/ratings')
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
views_table = soup.find('table', class_='views-table')
```

```python
rows = views_table.find('tbody').find_all('tr')
```

```python
titles = [row.find(class_='views-field-title')
             .get_text()
             .strip() 
          for row in rows]
```

```python
home_url = 'https://www.allsides.com'
url = [f"{home_url}{row.find(class_='views-field-title').find('a')['href']}" for row in rows]
```

```python
url
```

```python
def extract_rating(href):
    return href.split('/')[-1]

bias_ratings = [
    # tagi `a` w rodku tagu `td` o klasie `views-field-field-bias-image`
    extract_rating(
        row.select_one('td.views-field-field-bias-image a')['href']
    )
    for row in rows
]
```

```python
bias_ratings
```

```python
agrees = [
    int(row.find('span', class_='agree').get_text())
    for row in rows
]

disagrees = [
    int(row.find('span', class_='disagree').get_text())
    for row in rows
]
```

```python
agrees
```

```python
def get_agreeance_text(ratio):
    if ratio > 3: return "absolutely agrees"
    elif 2 < ratio <= 3: return "strongly agrees"
    elif 1.5 < ratio <= 2: return "agrees"
    elif 1 < ratio <= 1.5: return "somewhat agrees"
    elif ratio == 1: return "neutral"
    elif 0.67 < ratio < 1: return "somewhat disagrees"
    elif 0.5 < ratio <= 0.67: return "disagrees"
    elif 0.33 < ratio <= 0.5: return "strongly disagrees"
    elif ratio <= 0.33: return "absolutely disagrees"
    else: return None
    
print(get_agreeance_text(2.5))
```

```python
ratios = [agrees/disagrees for agrees, disagrees in list(zip(agrees, disagrees))]
commtexts = [
    get_agreeance_text(ratio)
    for ratio in ratios
]

commtexts
```

```python
df_site = pd.DataFrame({
    'title': titles,
    'url': url,
    'rating': bias_ratings,
    'agree': agrees,
    'disagree': disagrees,
    'ratio': ratios,
    'commtext': commtexts
})
```

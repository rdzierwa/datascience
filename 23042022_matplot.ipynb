### Kolorki

```python
colors = [
    'blue',
    'g',#rgbcmyk
    '0.75',#greyscal
    '#FFDD44',# hex
    (1,0.3,0.3), # rgb(x,y,z)
    'chartreuse'#html
]
```

```python
list(enumerate(colors))
```

```python
for i, color in enumerate(colors):
    plt.plot(x, np.sin(x-i), color = color)
```

```python
linestyles = [
    'solid', # ciagla
    'dashed', # kreskowa
    'dashdot', # kropka - kreska
    'dotted', # kropkowy
    '-',
    '--',
    '-.',
    ':'
]
```

```python
for i, linestyle in enumerate(linestyles):
    plt.plot(x, x+i, linestyle=linestyle)
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black', linestyle='dotted')
```

```python
rng = np.random.RandomState(0)
plt.figure(figsize=(10, 9), dpi=100)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label=f'marker={marker}')
plt.legend() # mona miesza marker + style lini w postaci np :o - kropkami polacz kropy bdce wynikiem na wykresieplt.plot(x, y, ':o', color='red');
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
features = iris.data.T
```

```python
iris.feature_names
```

```python
plt.figure(figsize=(10,9), dpi =110)
plt.scatter(
    features[0],
    features[1],
    alpha=0.2,
    s=100*features[3],
    c=iris.target,
    cmap='viridis'
)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
